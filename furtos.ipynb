```python
import pandas as pd
from pandas import Series, DataFrame
import folium
from folium import plugins
```

```python

# df = pd.read_csv(filename, delimiter= ';', encoding='latin-1')            
path = 'C:\\Users\Rafael\Desktop\Fatec\Sem_3\ED\jupyter/'
startName = 'furtos_'
filenames = ['maio','junho','julho','agosto', 'setembro']
extension = '.csv'
```

```python
#Função para percorrer os arquivos de dados e popular o dataframe.
def populaDataFrame():
    dfList = list()

    for file in filenames:
        df= pd.DataFrame(pd.read_csv(path+startName+file+extension, delimiter= ';', encoding='latin-1'))
        dfList.append(df)

    df = pd.concat(dfList)
    return df
```

```python
# df.sample(10)

df = populaDataFrame()
df.reset_index()
```

```python
df.groupby('PERIDOOCORRENCIA').size().sort_values().plot(kind='barh')
```

```python
df.groupby('DESCR_TIPO_VEICULO').size().sort_values().plot(kind='barh')
```

```python
#Criamos um novo dataframe para trabalhar com as informações que iremos utilizar
geolocalizacoes = df[['LATITUDE', 'LONGITUDE']]

#Processo de limpeza da base removendo os valores nulos
#Identifica a quantidade de resultados nulos existentes na base
geolocalizacoes.isnull().sum()

#Remove estes valores nulos e persiste na indexação
geolocalizacoes.dropna(inplace=True)
#Valida novamente a quantidade de resultados nulos
geolocalizacoes.isnull().sum()
```

```python
#Através da biblioteca do Folium criamos um novo mapa e populamos ele com as latitudes e longitudes do dataframe
coordenadas = []
lat = [ float(val.replace(',','.')) for val in geolocalizacoes['LATITUDE'].values]

long = [ float(val.replace(',','.')) for val in geolocalizacoes['LONGITUDE'].values]

mapa = folium.Map(location=[-23.1791, -45.8872],tiles='OpenStreetMap',zoom_start=12)
 
for la,lo in zip(lat,long):
    coordenadas.append([la,lo])
    
mapa.add_child(plugins.HeatMap(coordenadas))
```
